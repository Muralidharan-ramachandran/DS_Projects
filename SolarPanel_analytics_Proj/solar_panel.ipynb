```python
import tensorflow as tf
import numpy as np
from PIL import Image

# Load model
model = tf.keras.models.load_model("mobilenetv2_solar_F.keras")

# Define your class names
class_names = ['bird-drop', 'clean', 'dusty', 'electrical-damage', 'physical-damage', 'snow-covered']

# Preprocess function
def preprocess_image(image_path):
    img = Image.open(image_path).convert("RGB").resize((128, 128))
    img = np.array(img) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# Predict
img = preprocess_image("test_image2.jpg")
pred = model.predict(img)
predicted_class = class_names[np.argmax(pred)]
print("Predicted class:", predicted_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Class distribution (as provided)
class_counts = [214, 193, 190, 99, 69, 123]
plt.figure(figsize=(10, 6))
sns.barplot(x=class_names, y=class_counts)
plt.title("Class Distribution in Solar Panel Dataset")
plt.xticks(rotation=45)
plt.ylabel("Number of Images")
plt.savefig("class_distribution.png")
plt.show()
```
